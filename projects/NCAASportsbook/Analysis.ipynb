```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import date

num = 1
for i in range(num, 31):
    x = pd.read_pickle('Equation%s.pkl' % i)

    x = x[x['date'] != date.today()]
    x['Home_Spread'] = abs(x['Home_Spread'].astype(float))
    t = x[x['Hit'] == True]
    f = x[x['Hit'] == False]

    hit1 = t[t['Home_Spread'] <= 5]['Home_Spread'].values
    hit2 = t[(t['Home_Spread'] > 5) & (t['Home_Spread'] <= 10)]['Home_Spread'].values
    hit3 = t[(t['Home_Spread'] > 10) & (t['Home_Spread'] <= 15)]['Home_Spread'].values
    hit4 = t[(t['Home_Spread'] > 15) & (t['Home_Spread'] <= 20)]['Home_Spread'].values
    hit5 = t[(t['Home_Spread'] > 20) & (t['Home_Spread'] <= 25)]['Home_Spread'].values
    hit6 = t[(t['Home_Spread'] > 25) & (t['Home_Spread'] <= 30)]['Home_Spread'].values
    hit7 = t[(t['Home_Spread'] > 30) & (t['Home_Spread'] <= 35)]['Home_Spread'].values
    hit8 = t[t['Home_Spread'] > 35]['Home_Spread'].values

    miss1 = f[f['Home_Spread'] <= 5]['Home_Spread'].values
    miss2 = f[(f['Home_Spread'] > 5) & (f['Home_Spread'] <= 10)]['Home_Spread'].values
    miss3 = f[(f['Home_Spread'] > 10) & (f['Home_Spread'] <= 15)]['Home_Spread'].values
    miss4 = f[(f['Home_Spread'] > 15) & (f['Home_Spread'] <= 20)]['Home_Spread'].values
    miss5 = f[(f['Home_Spread'] > 20) & (f['Home_Spread'] <= 25)]['Home_Spread'].values
    miss6 = f[(f['Home_Spread'] > 25) & (f['Home_Spread'] <= 30)]['Home_Spread'].values
    miss7 = f[(f['Home_Spread'] > 30) & (f['Home_Spread'] <= 35)]['Home_Spread'].values
    miss8 = f[f['Home_Spread'] > 35]['Home_Spread'].values

    labels = ['< 5', '5 <= 10', '10 <= 15', '15 <= 20', '20 <= 25', '25 <= 30', '30 <= 35', '> 35']
    successRate = {'Hit': (len(hit1), len(hit2), len(hit3), len(hit4), len(hit5), len(hit6), len(hit7), len(hit8)), 
                'Miss': (len(miss1), len(miss2), len(miss3), len(miss4), len(miss5), len(miss6), len(miss7), len(miss8))}
    colors = {'Hit':'green', 'Miss':'red'}

    labelLocations = np.arange(len(labels))
    width = 0.25
    multiplier = 0

    plt.bar(labelLocations, successRate.get('Hit'), color = 'green', width = width, edgecolor = 'black', label = 'Hit')
    plt.bar(labelLocations + width, successRate.get('Miss'), color = 'red', width = width, edgecolor = 'black', label = 'Miss')
    plt.xlabel("Spread Margin")
    plt.ylabel("Success Rate")
    plt.title('Equation %s: Success Rate vs Spread Margin' % i)
    plt.xticks(labelLocations + (width/2), labels, rotation = 60)
    plt.legend()

    plt.show()
```
